## 1. Mean Absolute Error (MAE)
  The Mean Absolute Error is the average of the absolute differences between the predicted values and the actual target values. It measures the magnitude of errors in a set of predictions without considering their direction.


## 2. Mean Squared Error (MSE)
  The Mean Squared Error is the average of the squared differences between the predicted values and the actual target values. It gives more weight to larger errors due to squaring, making it sensitive to outliers.

## 3. Accuracy
  Accuracy is the ratio of correctly predicted observations to the total observations. It measures how often the model makes correct predictions overall.

---

```python
display(Image(filename="images/acc.png"))
```

## 4. Precision
- **Definition**:  
  True Positives are the cases where the model correctly predicts the positive class. In other words, when the actual value is positive and the model also predicts it as positive. This means the model has made a correct positive prediction.

  False Positives are the cases where the model incorrectly predicts the positive class. This happens when the actual value is negative, but the model wrongly classifies it as positive.

  Precision is the ratio of correctly predicted positive observations to the total predicted positive observations. It measures how many of the predicted positives are actually correct.

# **Evaluation Metrics and Model Resources**

## **Regression Model Metrics**
1. [Root Mean Squared Error (RMSE)](https://scikit-learn.org/1.5/modules/generated/sklearn.metrics.root_mean_squared_error.html)  
   Measures the square root of the average squared differences between predicted and actual values.
2. [Mean Absolute Error (MAE)](https://scikit-learn.org/1.5/modules/generated/sklearn.metrics.mean_absolute_error.html)  
   Represents the average magnitude of errors in a set of predictions, without considering their direction.
3. [Mean Squared Error (MSE)](https://scikit-learn.org/1.5/modules/generated/sklearn.metrics.mean_squared_error.html)  
   Measures the average squared differences between predicted and actual values.
4. [Mean Absolute Percentage Error (MAPE)](https://scikit-learn.org/1.5/modules/generated/sklearn.metrics.mean_absolute_percentage_error.html)  
   Represents the average percentage error between predicted and actual values.

---

## **Classification Model Metrics**
1. [Accuracy Score](https://scikit-learn.org/1.5/modules/generated/sklearn.metrics.accuracy_score.html)  
   The ratio of correctly predicted observations to the total observations.
2. [Precision Score](https://scikit-learn.org/1.5/modules/generated/sklearn.metrics.precision_score.html)  
   The ratio of true positives to the sum of true positives and false positives (useful for class imbalance).
3. [Recall Score](https://scikit-learn.org/1.5/modules/generated/sklearn.metrics.recall_score.html)  
   The ratio of true positives to the sum of true positives and false negatives (sensitivity or true positive rate).
4. [F1 Score](https://scikit-learn.org/1.5/modules/generated/sklearn.metrics.f1_score.html)  
   The harmonic mean of precision and recall, balancing the two metrics.
5. [ROC AUC Score](https://scikit-learn.org/1.5/modules/generated/sklearn.metrics.roc_auc_score.html)  
   Measures the area under the Receiver Operating Characteristic (ROC) curve.
